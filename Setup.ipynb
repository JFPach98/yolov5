# Setup

```python
import os
import shutil
import numpy as np
import random
import itertools
```

Se obtiene una lista de las im√°genes y de las labels.

```python
labels = os.listdir('./labels')
images = os.listdir('./dataset')
```

Se randomiza la lista 'labels' para crear el dataset de entrenamiento y el de pruebas. Se hace un split de 75-25 % para los datasets.

```python
train_labels = []
test_labels = []

randomized = labels[:]
random.shuffle(labels)

quarter = len(labels)//4
quarter

train_labels = labels[quarter:]
test_labels = labels[:quarter]
```

```python
train_labels
```

Se crea el dataset de entrenamiento, con las direcciones de las im√°genes.

```python
train_images = []

for x in train_labels:
    image = x.replace('.txt', '.JPG')
    train_images.append(image)

train_images
```

Y el dataset de prueba.

```python
test_images = []

for x in test_labels:
    image = x.replace('.txt', '.JPG')
    test_images.append(image)

test_images
```

Se mueven los datasets en a su respectiva carpeta:

```python
data = os.listdir('./dataset')
lab = os.listdir('./lab')
for (x, y) in zip(data, lab):
    shutil.move('./dataset/' + x, './testing/test/images/')
    shutil.move('./lab/' + y, './testing/test/labels/' )
```

```python


for (x, y) in zip(train_images, train_labels):
    shutil.move('./dataset/' + x, './data/train/images/')
    shutil.move('./labels/' + y, './data/train/labels/' )
for (x, y) in zip(test_images, test_labels):
    shutil.move('./dataset/' + x, './data/test/images/')
    shutil.move('./labels/' + y, './data/test/labels/' )
```
